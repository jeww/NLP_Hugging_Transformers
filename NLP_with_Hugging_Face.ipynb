```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "AI is transforming the future of technology and business.",
    candidate_labels=["technology", "health", "education", "business"],
)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")

classifier(
    "The government has announced new healthcare policies.",
    candidate_labels=["health", "politics", "sports"],
)
```

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
result = classifier(
    "Sustainable energy solutions are essential for combating climate change.",
    candidate_labels=['environment', 'technology', 'health', 'politics'],
)

print(result)
```

```python
from transformers import pipeline

unmasker = pipeline("fill-mask")
unmasker("This course will teach you all about <mask> models.", top_k=2)
```

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)

ner("Elon Musk is the CEO of SpaceX and Tesla.")
```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Matahari adalah bintang pusat sistem tata surya yang terdiri dari plasma dan menghasilkan energi melalui fusi nuklir. Ia adalah sumber utama cahaya dan energi untuk Bumi, yang memungkinkan kehidupan di planet ini. Diameter Matahari sekitar 1,4 juta kilometer, dan terletak sekitar 150 juta kilometer dari Bumi."
question = "Berapa diameter Matahari?"

result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")

```

```python
from transformers import pipeline

qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")
context = "Soekarno adalah Presiden Republik Indonesia pertama, yang lahir pada 6 Juni 1901 dan meninggal pada 21 Juni 1970. Ia dikenal sebagai Bapak Bangsa dan salah satu proklamator kemerdekaan Indonesia. Soekarno berperan penting dalam pengembangan ideologi Pancasila sebagai dasar negara."
question = "Siapa itu Soekarno?"

result = qa_pipeline(question=question, context=context)
print(f"Jawaban: {result['answer']}")

```

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
result = classifier("This workshop exceeded my expectations! I can't wait to explore more options.")

print(result)
```

```python
from transformers import pipeline

summarizer = pipeline("summarization")


text_to_summarize = """
    The global economy has experienced significant shifts in recent years.
    Many nations are focusing on sustainable development and the transition
    to renewable energy sources. This shift is driven by the urgent need
    to combat climate change and reduce dependency on fossil fuels.
    Countries like Germany and Denmark are leading the way in wind
    energy, while others invest heavily in solar power technologies.

    Despite these advancements, challenges remain, including the need
    for improved energy storage solutions and grid infrastructure.
    Emerging economies, especially in Asia, are also adapting to these
    changes, seeking to balance growth with environmental responsibility.
    The collaboration between nations is crucial for achieving global
    sustainability goals and ensuring a cleaner planet for future generations.
"""


summary = summarizer(text_to_summarize)

print("Summary:")
print(summary[0]['summary_text'])
```

```python
from transformers import pipeline

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-id-en")


text_to_translate = "Saya menikmati membaca buku setiap malam."
result = translator(text_to_translate)

print(result[0]['translation_text'])
```

# Analisis

**zero-shot-classification :** Sangat berguna untuk pengkategorian cepat, tetapi akurasinya bisa bervariasi tergantung konteks yang tidak selalu tepat.

**text-generation:** Hasil yang dihasilkan bisa jadi tidak konsisten, terkadang tidak relevan dengan input.

**fill-mask :** Walaupun membantu, model terkadang mengisi dengan kata yang kurang tepat, mengurangi makna asli.

**ner:** Mampu mengekstrak informasi penting, namun bisa salah mengenali entitas dalam konteks tertentu.

**question-answering:** Menarik dan interaktif, tetapi jawaban yang diberikan tidak selalu akurat atau memadai.

**sentiment-analysis:** Berguna untuk memahami perasaan, namun bisa salah menafsirkan nada atau konteks.

**summarization:** Memudahkan pemahaman, tetapi terkadang menyederhanakan terlalu banyak sehingga kehilangan informasi penting.

**translation:** Memudahkan komunikasi antarbahasa, namun hasil terjemahan bisa kurang alami dan mengandung kesalahan.
